```python
import utils
import pandas as pd
from itables import show
from constants import FA_INDICATORS
```

```python
sp500 = utils.spy_components()
data = utils.toDF(sp500)
hk_data = utils.toDF(['10']
```

```python
compact_data = data[['marketCap', 'forwardPE', 'P/B', 'ROE', 'trailingPegRatio', 'trailingDivYield']]
```

```python
def my_show_itable(df):
    show(df.reset_index(), 
         layout = {"top1": "searchBuilder"},
         searchBuilder={
            "preDefined": {
                "criteria": [
                    {"data": "ROE", "condition": ">", "value": [0.1,]}, 
                    {"data": "P/B", "condition": "<", "value": [10,]}, 
                ]
            }
        },
         buttons = [
            "pageLength",
            "copyHtml5", "csvHtml5", "excelHtml5", 
            {
                "extend": 'colvis',
                "collectionLayout": 'fixed columns',
                "popoverTitle": 'Column visibility control'
            },
         ]
    )
#
my_show_itable(compact_data)
```

```python
top_2_by_sector = (
    compact_data.groupby('sectorKey', group_keys=False)
      .apply(lambda x: x.sort_values('marketCap', ascending=False).head(2))
)
show(top_2_by_sector)
```
